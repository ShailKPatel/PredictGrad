```python
import os

EXCLUDE_DIRS = {"__pycache__", ".venv", ".git"}
EXCLUDE_FILES = {".DS_Store"}
MAX_DEPTH = 3  # Adjust depth as needed

def print_tree(start_path, prefix="", depth=0):
    if depth > MAX_DEPTH:
        return

    entries = sorted(os.listdir(start_path))
    entries = [e for e in entries if e not in EXCLUDE_FILES]

    for i, entry in enumerate(entries):
        path = os.path.join(start_path, entry)
        is_last = i == len(entries) - 1
        connector = "└── " if is_last else "├── "

        if os.path.isdir(path):
            if entry in EXCLUDE_DIRS:
                continue
            print(f"{prefix}{connector}{entry}/")
            extension = "    " if is_last else "│   "
            print_tree(path, prefix + extension, depth + 1)
        else:
            print(f"{prefix}{connector}{entry}")

# Run the function from the desired root (e.g., current dir)
print_tree(".")
```
